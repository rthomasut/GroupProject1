```python
import pandas as pd
import numpy as np
import datetime as dt
import json

path = "WeatherAPIData.csv"
df = pd.read_csv(path)
df.head()
```

```python
df['Date'] = df['Date'].str.strip('[]').astype('datetime64[ns]')
df['Weathercode'] = df['Weathercode'].str.strip('[]').astype(int)
df['Max Temperature'] = df['Max Temperature'].str.strip('[]').astype(float)
df['Min Temperature'] = df['Min Temperature'].str.strip('[]').astype(float)
df['Mean Temperature'] = df['Mean Temperature'].str.strip('[]').astype(float)
df['Sunrise'] = df['Sunrise'].str.strip('['']').astype('datetime64[ns]')
df['Sunrise'] = pd.to_datetime(df['Sunrise'], format='%H:%M').dt.time
df['Sunset'] = df['Sunset'].str.strip('['']').astype('datetime64[ns]')
df['Sunset'] = pd.to_datetime(df['Sunset'], format='%H:%M').dt.time
df['Total Precipitation'] = df['Total Precipitation'].str.strip('[]').astype(float)
df['Total Rain'] = df['Total Rain'].str.strip('[]').astype(float)
df['Total Snow'] = df['Total Snow'].str.strip('[]').astype(float)
df.head()
```
